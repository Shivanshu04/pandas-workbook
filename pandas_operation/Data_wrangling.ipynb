```python
# merge the two dataframes using right join
right_merge_df =pd.merge(english_mov,#left Df
                    long_mov,# right df
                    how='right', #type of join
                    left_on ='id',
                    right_on='id'#joining key
                    )
```

```python
print(right_merge_df.shape)
print(long_mov.shape)
```
