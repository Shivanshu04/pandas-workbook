```python
#ndim no of dimensiom
import numpy as np
prim_array = np.array([[2,3,5,7,9],[1,2,3,4,5]])
print("Dimension of prim_array is =" ,prim_array.ndim)
```

```python
#ndim no of dimensiom
import numpy as np
prim_array = np.array([2,3,5,7,9,7,7])
print("Dimension of prim_array is =" ,prim_array.ndim)
```

```python
#ndim no of dimensiom
import numpy as np
prim_array = np.array([[2,3,5,7,9,7,7],[3,4,5,6,8,9,9]])
print("Dimension of prim_array is =" ,prim_array.ndim)
```

```python
#ndim no of dimensiom
import numpy as np
prim_array = np.array([[2,3,5,7,9],[1,2,3,4,5]])
print("Data type of prim_array is =" ,prim_array.dtype)
```

```python
#ndim no of dimensiom
import numpy as np
prim_array = np.array([[2,3,5,7,9],[1,2,3,4,5]])
print("Data type of prim_array is =" ,prim_array.astype)
```
